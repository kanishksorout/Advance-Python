```python
!pip install numpy pandas matplotlib


# Assignment 1  Working with NumPy
import numpy as np

# 1. 1D NumPy Array Operations
arr_1d = np.arange(1, 21)
print("Array:", arr_1d)
print("Sum:", np.sum(arr_1d))
print("Mean:", np.mean(arr_1d))
print("Median:", np.median(arr_1d))
print("Standard Deviation:", np.std(arr_1d))
print("Indices > 10:", np.where(arr_1d > 10))

# 2. 2D NumPy Array Operations
arr_2d = np.arange(1, 17).reshape(4, 4)
print("\n2D Array:\n", arr_2d)
print("Transpose:\n", arr_2d.T)
print("Row-wise sum:", np.sum(arr_2d, axis=1))
print("Column-wise sum:", np.sum(arr_2d, axis=0))

# 3. Two 3x3 Random Arrays
arr_a = np.random.randint(1, 21, (3, 3))
arr_b = np.random.randint(1, 21, (3, 3))
print("\nArray A:\n", arr_a)
print("Array B:\n", arr_b)
print("Addition:\n", arr_a + arr_b)
print("Subtraction:\n", arr_a - arr_b)
print("Multiplication:\n", arr_a * arr_b)
print("Dot Product:\n", np.dot(arr_a, arr_b))

# 4. Reshape and Slice
arr_reshaped = np.arange(1, 13).reshape(3, 4)
print("\nReshaped Array:\n", arr_reshaped)
print("First 2 Rows, Last 2 Columns:\n", arr_reshaped[:2, -2:])

# -------------------------------------------------------------------

# Assignment 2  Working with Pandas
import pandas as pd

# 1. Create and Explore DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
print("\nSummary statistics (Age & Salary):\n", df[['Age', 'Salary']].describe())
print("Average HR Salary:", df[df['Department'] == 'HR']['Salary'].mean())

# 2. Add Bonus Column
df['Bonus'] = df['Salary'] * 0.10
print("\nDataFrame with Bonus:\n", df)

# 3. Filter Age Between 25 and 30
print("\nEmployees aged 25 to 30:\n", df[(df['Age'] >= 25) & (df['Age'] <= 30)])

# 4. Group by Department and Average Salary
print("\nAverage salary by department:\n", df.groupby('Department')['Salary'].mean())

# 5. Sort by Salary and Save to CSV
sorted_df = df.sort_values(by='Salary')
print("\nSorted by Salary:\n", sorted_df)
# Uncomment below line to save
# sorted_df.to_csv('sorted_employees.csv', index=False)

# -------------------------------------------------------------------

# Assignment 3  Working with Matplotlib
import matplotlib.pyplot as plt

# 1. Line Plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o')
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

# 2. Bar Graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.figure(figsize=(6, 4))
plt.bar(students, marks, color=['blue', 'green', 'red', 'purple'])
plt.title("Marks of Students")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()

# 3. Pie Chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1 if r == max(revenue) else 0 for r in revenue]
plt.figure(figsize=(6, 6))
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title("Revenue Distribution")
plt.show()

# 4. Histogram
rand_nums = np.random.randint(1, 101, 1000)
plt.figure(figsize=(6, 4))
plt.hist(rand_nums, bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
